# Feature derivation:

```python
# hours before launching
df['hbl'] = df.apply(lambda x: (x.launched_at - x.created_at)/3600 , axis=1)
```

```python

df['log_hbl']=df['hbl'].apply(lambda x: np.log(x))
sns.distplot(df.log_hbl, fit = norm)
```

```python
stats.probplot(df.log_hbl, plot=plt)
```

```python
# hours before deadline 
df['hbd'] = df.apply(lambda x: (x.deadline - x.launched_at)/3600 , axis=1)
```

# Modelling:

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn import metrics
```

```python
df1 = pd.read_csv('clean_dataset.csv')
```

```python
df1
```

```python
x_train, x_test, y_train, y_test = train_test_split(df1.blurb, df1.funded, test_size=0.2)
```

```python
text_clf_nb = Pipeline([
    ('tfidf', TfidfVectorizer()),
    ('clf', MultinomialNB()),
])
```

```python
text_clf_nb.fit(x_train, y_train)
```

```python
def plot_roc(model, x, y, name='linear'):
    y_text = model.predict_proba(x)

    curve_linear = metrics.roc_curve(y, y_text[:, 1])

    # Compute AUC from the created ROC
    auc_linear = metrics.auc(curve_linear[0], curve_linear[1])

    plt.plot(curve_linear[0], curve_linear[1], label=f'{name} (area = %0.2f)' % auc_linear)

    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.0])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC curve');

    plt.legend();
```

```python
plot_roc(text_clf_nb, x_test, y_test, 'tdidf bayes')
```

```python
def to_boolean(values: list):
    if values[0] > values[1]:
        return False
    else:
        return True

predicted = text_clf_nb.predict_proba(x_test)

predicted
result = []
for x in predicted:
    result.append(to_boolean(x))
    
metrics.accuracy_score(y_test, result)
```
