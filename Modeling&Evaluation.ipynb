# Baseline Model

```python
# the test data has 175 countries while the training data has 174, hence we add one extra category to the training data consisting of 0's
df['country_175'] = 0
```

```python
# we scale the numerical data, i.e. we scale all of the numbers between 0 and 1
df_scaled = clean.scale_data(df, exclude=['blurb', 'sent_blurb', 'name', 'hbl', 'goal', 'funded'])
```

```python
x_train, x_test, y_train, y_test = train_test_split(df_scaled, df.funded, test_size=0.3)
```

```python
clf = MultinomialNB()
```

```python
clf.fit(x_train, y_train)
```

```python
def plot_roc(model, x, y, name='linear'):
    y_text = model.predict_proba(x)

    curve_linear = metrics.roc_curve(y, y_text[:, 1])

    # Compute AUC from the created ROC
    auc_linear = metrics.auc(curve_linear[0], curve_linear[1])

    plt.plot(curve_linear[0], curve_linear[1], label=f'{name} (area = %0.2f)' % auc_linear)

    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.0])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC curve');

    plt.legend();
```

```python
plot_roc(clf, x_test, y_test, 'Multinomial NB')
```

##### Cross-validation

```python
scores = cross_val_score(clf, df_scaled, df.funded, cv=5)
predicted = cross_val_predict(clf, df_scaled, df.funded, cv=5)
print(metrics.accuracy_score(df.funded, predicted))
print(scores.mean(), scores)
# scores: 0.7219899163698932 when cv=100
```

```python
predicted = clf.predict(x_test)
metrics.accuracy_score(y_test, predicted)
```

### Hyper-parameters tuning using Grid Search:

```python
# Prepare parameters 
model.get_params()
#  Number of trees in random forst
n_estimators = [int(x) for x in np.linspace(start = 10, stop = 200, num = 20)]
# Number of features to consider at every split
max_features = ['sqrt', 'log2']
max_features.append(None)
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(10, 25, num = 5)]
max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]
# Method of selecting samples for training each tree
bootstrap = [True, False]

params = {'n_estimators': n_estimators,
          'max_features': max_features,
          'max_depth': max_depth,
          'min_samples_split': min_samples_split,
          'min_samples_leaf': min_samples_leaf,
          'bootstrap': bootstrap,
          }
```

```python
from sklearn.model_selection import RandomizedSearchCV
x_train, x_test, y_train, y_test = train_test_split(df_scaled, df.funded, test_size=0.3)
rf = RandomForestClassifier()
rf_random = RandomizedSearchCV(estimator = rf, 
                               param_distributions = params, 
                               n_iter = 10, 
                               n_jobs = -4, 
                               cv = 5) 
```

```python
rf_random.fit(x_train, y_train)
```

## Feature Impotance

```python
import time
from sklearn.inspection import permutation_importance
```

```python
start_time = time.time()
importances = best_model.feature_importances_
std = np.std([
    tree.feature_importances_ for tree in best_model.estimators_], axis=0)
elapsed_time = time.time() - start_time

print(f"Elapsed time to compute the importances: "
      f"{elapsed_time:.3f} seconds")
```

```python
# we get the indices of the 10 most important features
indx = np.argpartition(best_model.feature_importances_, -10)[-10:]
```

```python
df_copy = df.drop(['blurb', 'name', 'hbl', 'goal', 'funded'],axis=1)
```

```python
f_names = df_copy.iloc[:, indx].columns
```

```python
feature_names = [f'feature {i}' for i in [x for x in f_names]]
forest_importances = pd.Series(importances[indx], index=feature_names)

fig, ax = plt.subplots()
forest_importances.plot.bar(yerr=std[indx], ax=ax)
ax.set_title("Feature importances using MDI") # mean decrease in impurity
ax.set_ylabel("Mean decrease in impurity")
fig.tight_layout()
```
